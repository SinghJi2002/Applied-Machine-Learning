# Fruit Detection Model-Classification Model

```python
#First we import the neccessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#Lets import our dataset
fruits=pd.read_table('assets/fruit_data_with_colors.txt')
fruits.head()
```

## Lets seperate the training and testing data

```python
#Lets create a split between the dataset. One part will contain features that are used to train the dataset, while the other
#part will contain features that will be used to test the dataset.
x=fruits[['mass','width','height']]
y=fruits['fruit_label']

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

## Lets visualize the data at our disposal to gain important insights

```python
#This general plotation is done in order to establish relation between the suppossed input labels to run mutiple checks before 
#actually implementing the machine learning algorithm.
from matplotlib import cm
cmap=cm.get_cmap('gnuplot')
scatter=pd.plotting.scatter_matrix(x_train,c=y_train,marker='0',s=40,cmap=cmap,figsize=(12,12),hist_kwds={'bins':15})
```

## The Prediction Task

```python
#This whole section is based on skikit-learn module. Note the codes carefully here.
#Step 1: We first create our classifier/training object.

from sklearn.neighbors import KNeighborsClassifier
kNN=KNeighborsClassifier(n_neighbors=5)

#kNN is the classifier here that will actually implement the classification process via the kNN algorithm. Here we are using 
#a subattribute, neighbors, the set the value of k of KNN algo.
```

```python
#Step2: Train the classifier using the training set.
kNN.fit(x_train,y_train)
type(kNN)
```

```python
#Step3: Now our model has been trained on the basis of training set data. Now the next step is to use this trained model
#to make predictions. The code is as follows.

fruit_prediction = kNN.predict([[20, 4.3, 5.5]])

print(fruit_prediction)

#As observable we use the the predict method, to generate predictions from pre-trained models. Note the input in the predict 
#though. First of all its an array. Second of all, it has as many arguments as the features used in training the set.

#Here the 2 output represents the fruit labelled by label 2, that is lemon.
```

```python
#Step4: The next step in the process is to check the accuracy of our model. This is done via the score attribute.
kNN.score(x_train,y_train)

#As observable these work with 79.5 percent accuracy.
```
